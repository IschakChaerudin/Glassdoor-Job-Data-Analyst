```python
# Data Cleaning
```

```python
#Salary Estimate
```

```python
data['Salary Estimate(K)'].dtype
```

```python
data['Salary Estimate(K)'].value_counts()
```

```python
salary = data['Salary Estimate(K)'].apply(lambda x : x.split('(')[0])
```

```python
salary
```

```python
salary = salary.apply(lambda y : y.replace('$','').replace('K',''))
```

```python
salary
```

```python
data['min_salary'] = salary.apply(lambda t : t.split('-')[0])
```

```python
data['min_salary']
```

```python
data['max_salary'] = salary.apply(lambda t : t.split('-')[1])
```

```python
data['max_salary']
```

```python
data.head()
```

```python
data['min_salary'].unique()
```

```python
data['max_salary'].unique()
```

```python
data.info()
```

```python
data['min_salary'] = data['min_salary'].astype(int)
```

```python
data['max_salary'] = data['max_salary'].astype(int)
```

```python
data.info()
```

```python
data['avg_salary'] = (data['min_salary']+data['max_salary'])/2
```

```python
data.head()
```

```python
data['avg_salary'].dtype
```
